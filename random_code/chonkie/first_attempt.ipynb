```python
from chonkie import Pipeline

# Build and execute pipeline
doc = (Pipeline()
    .fetch_from("file", path="/Users/ksulahian/ML_journey/Karapathy/GPT_from_scratch/input.txt")
    .process_with("text")
    .chunk_with("recursive", chunk_size=512)
    # Add overlap context
    .refine_with("overlap", context_size=100, method="prefix")
    # Add embeddings
    # .refine_with("embedding", model="text-embedding-3-small")
    .run())

# Access chunks
print(f"Created {len(doc.chunks)} chunks")
for chunk in doc.chunks:
    print(f"{chunk.text}\n \n------------")
    break
```

```python
# Chunk with embeddings
doc = (Pipeline()
    .fetch_from("file", path="/Users/ksulahian/ML_journey/Karapathy/GPT_from_scratch/input.txt")
    .process_with("text")
    .chunk_with("semantic",
                threshold=0.8,
                chunk_size=1024,
                similarity_window=3)
    .refine_with("overlap", context_size=100)
    .refine_with("embedding", model="minishlab/potion-base-32M")
    .run())

# All chunks now have embeddings
for chunk in doc.chunks:
    if chunk.embedding is not None:
        print(f"Chunk: {chunk.text[:30]}... | Embedding shape: {chunk.embedding.shape}")
```

```python
import sys
import subprocess

# --- 1. Check where your executable is ---
print(f"Python Executable: {sys.executable}")
print("---")

# --- 2. Check all installed packages ---
try:
    # Use pip list to see what's in this *exact* environment
    result = subprocess.run([sys.executable, '-m', 'pip', 'list'], capture_output=True, text=True, check=True)
    
    # Check for chonkie and its dependencies
    packages = result.stdout
    print("Checking for 'chonkie' and 'sentence-transformers'...")
    
    found_chonkie = False
    found_transformers = False
    
    for line in packages.split('\n'):
        if 'chonkie' in line:
            print(f"FOUND: {line}")
            found_chonkie = True
        if 'sentence-transformers' in line:
            print(f"FOUND: {line}")
            found_transformers = True
            
    if not found_chonkie:
        print("ERROR: 'chonkie' is NOT installed in this environment.")
    if not found_transformers:
        print("ERROR: 'sentence-transformers' (part of [semantic]) is NOT installed in this environment.")
        
    print("---")
    print(packages)

except Exception as e:
    print(f"An error occurred while running pip list: {e}")
```
